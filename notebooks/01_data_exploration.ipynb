```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/simulated/longitudinal_simulated.csv")

sample_ids = (
    df["participant_id"]
    .drop_duplicates()
    .sample(10, random_state=1)
)

for pid in sample_ids:
    sub = df[df["participant_id"] == pid]
    plt.plot(sub["visit_month"], sub["memory"], marker="o")

plt.xlabel("Months")
plt.ylabel("Memory score")
plt.title("Simulated memory trajectories")
plt.show()

```
