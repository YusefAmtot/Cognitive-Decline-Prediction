```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/simulated/longitudinal_simulated.csv")

# pick a small number so the legend stays readable
sample_ids = df["participant_id"].drop_duplicates().sample(5, random_state=0)

plt.figure(figsize=(8, 5))

for pid in sample_ids:
    sub = df[df["participant_id"] == pid]

    # age is the same for all visits for this person
    age = sub["age_baseline"].iloc[0]

    plt.plot(
        sub["visit_month"],
        sub["memory"],
        marker="o",
        label=f"Age {int(age)}"
    )

plt.xlabel("Visit month")
plt.ylabel("Memory score")
plt.title("Simulated memory trajectories (labelled by age)")
plt.legend(title="Participant age")
plt.grid(True)
plt.show()


```
